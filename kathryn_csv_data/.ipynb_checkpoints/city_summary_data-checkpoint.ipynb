```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import seaborn as sns
%matplotlib inline
```

```python
wuhan = pd.read_csv("Wuhan/output_data/wuhan_summary_data.csv")
wuhan = wuhan.set_index("Month")
wuhan
```

```python
rome= pd.read_csv("Guido/output_data/guido_summary_data.csv")
rome = rome.set_index("Month")
rome
```

```python
stockholm = pd.read_csv("Stockholm/output_data/stockholm_summary_data.csv")
stockholm = stockholm.set_index("Month")
stockholm
```

```python
melbourne = pd.read_csv("Melbourne/output_data/melbourne_summary_data.csv")
melbourne = melbourne.set_index("Month")
melbourne
```

```python
wgsm_summary = pd.concat([wuhan,rome,stockholm,melbourne], axis=0,keys=['Wuhan','Rome','Stockholm','Melbourne'])
wgsm_summary
```

```python
wgsm_summary.to_csv("city_summary_data_results.csv")
```
