```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import seaborn as sns
%matplotlib inline
```

```python
testing = pd.read_csv("city_summary_data_results.csv")
testing
```

```python
testing2 = testing.rename(columns={"Unnamed: 0":"City"})
testing2
```

```python
summary = testing2.groupby(["City"])
```

```python
avg_2018 = round(summary["Avg AQI_18"].mean(),2)
avg_2019 = round(summary["Avg AQI_19"].mean(),2)
avg_2020 = round(summary["Avg AQI_20"].mean(),2)
```

```python
avg_aqi_cities = pd.DataFrame({
                            "Avg AQI 2018":avg_2018,
                            "Avg AQI 2019":avg_2019,
                            "Avg AQI 2020":avg_2020
    
    
                        })
```

```python
cities_df = avg_aqi_cities.reset_index()
```

```python
cities_df
```

```python
import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from api_keys import g_key
```

```python
cities_df["Lat"]=""
cities_df["Lng"]=""
```

```python
cities_df
```

```python
params = {"key":g_key}

for index, row in cities_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    city = row["City"]
    
    
    params['address'] = f"{city}"
    
    cities_lat_lng = requests.get(base_url,params=params)
    
    cities_lat_lng = cities_lat_lng.json()
    
    cities_df.loc[index, "Lat"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
    cities_df.loc[index, "Lng"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]
  
cities_df
```

```python
import gmaps

# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
locations = cities_df[["Lat","Lng"]]
# aqi_15 = cities["Avg AQI 2015"].astype(float)
# aqi_16 = cities["Avg AQI 2016"].astype(float)
# aqi_17 = cities["Avg AQI 2017"].astype(float)
# aqi_18 = cities["Avg AQI 2018"].astype(float)
# aqi_19 = cities["Avg AQI 2019"].astype(float)   
aqi_20 = cities_df["Avg AQI 2020"].astype(float)
```

```python
fig = gmaps.figure(map_type="TERRAIN")

heat_layer = gmaps.heatmap_layer(locations, weights=aqi_20,
                                dissipating=False,max_intensity=100,
                                point_radius=10)

info_box_template = """
<dl>
<dt>City</dt><dd>{City}</dd>
<dt>AQI<dt><dd>{Avg AQI 2020}</dd>
</dl>
"""

city_info = [info_box_template.format(**row) for index, row in cities_df.iterrows()]
locations = cities_df[["Lat","Lng"]] 

symbols = gmaps.symbol_layer(locations, fill_color='red', stroke_color='red',info_box_content=city_info)

                         
                         
fig.add_layer(heat_layer)
fig.add_layer(symbols)
fig
```
