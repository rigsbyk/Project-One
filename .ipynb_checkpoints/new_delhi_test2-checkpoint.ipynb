```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mumbai = "../AQI CSV Files/Mumbai_csv.csv"
new_delhi = "../AQI CSV Files/New_delhi_csv.csv"
```

```python
mumbai = pd.read_csv(mumbai, delimiter=", ")
new_delhi = pd.read_csv(new_delhi, delimiter=", ")
```

```python
new_delhi.info()
```

```python
new_delhi[["Year","Month", "Day"]] = new_delhi["date"].str.split("/", n = 2, expand = True) 
new_delhi.drop(columns =["date"], inplace = True) 
```

```python
new_delhi["Year"] = new_delhi["Year"].astype(int)
new_delhi["Month"] = new_delhi["Month"].astype(int)
new_delhi = new_delhi[new_delhi["Year"]>2017]
new_delhi = new_delhi[new_delhi["Month"]<8]
new_delhi
```

```python
new_delhi = new_delhi[["Year", "Month", "Day", "pm25"]]
new_delhi.rename(columns={"pm25":"AQI"},inplace=True)
new_delhi
```

```python
new_delhi.sort_values(by=['Year','Month'], ascending=True)
new_delhi
```

```python
# sorted_new_delhi = new_delhi.groupby(["Year","Month"]).agg({"AQI":["min","max","mean"]})
```

```python
new_delhi_2018 = new_delhi[new_delhi["Year"]==2018]
new_delhi_2018
```

```python
sorted_new_delhi = new_delhi_2018.groupby(["Month"]).agg({"AQI":["min","max","mean"]})
sorted_new_delhi_2018
```

```python
sorted_new_delhi_2018.plot.line()
```

```python
full_new_delhi=new_delhi.groupby(["Year","Month"]).agg({"AQI":["mean"]}).unstack()
full_new_delhi
```

```python
fig, ax=plt.subplots(figsize=(15,7))
full_new_delhi.iloc[0].plot(ax=ax)
full_new_delhi.iloc[1].plot(ax=ax)
full_new_delhi.iloc[2].plot(ax=ax)
plt.legend()
```

```python
min_max_new_delhi=new_delhi.groupby(["Year","Month"]).agg({"AQI":["min","max"]})
fig, ax=plt.subplots(figsize=(15,7))
min_max_new_delhi.loc[2018].plot.line(ax=ax, label="2018")
min_max_new_delhi.loc[2019].plot.line(ax=ax)
min_max_new_delhi.loc[2020].plot.line(ax=ax)
```

```python
test = new_delhi.groupby(["Year","Month"]).agg({"AQI":["mean","min","max"]})
test.unstack(level=0)
```
